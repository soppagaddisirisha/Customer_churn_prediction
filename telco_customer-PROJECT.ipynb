```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ","_")
```

```python
target = "churn"
x = df.drop(columns=[target])
y = df[target]
```

```python
print(df.columns)
```

```python
numeric_cols = x.select_dtypes(include=['int64','float64']).columns
categorical_cols = x.select_dtypes(include=['object','bool']).columns
numeric_cols,categorical_cols
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
preprocess = ColumnTransformer(transformers=[("num",StandardScaler(),numeric_cols),("cat",OneHotEncoder(handle_unknown='ignore'),categorical_cols)])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
model = Pipeline(steps=[("preprocess",preprocess),("classifier",LogisticRegression(max_iter=1000))])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42, stratify=y)
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
pred = model.predict(x_test)
print("Acuuracy:",accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```
